# Queue Position Estimation Algorithms
Models average order queue execution paths comparing FIFO (First-in-first-out) allocation with Pro-Rata configurations.


```python
queue_depth = 1200
order_size = 100
incoming_fills = [300, 400, 500, 300]

position = queue_depth
for i, fill in enumerate(incoming_fills):
    if position > fill:
        position -= fill
        print(f"Fill {i+1} size: {fill} | Remaining ahead in queue: {position}")
    else:
        executed = order_size - (position + order_size - fill)
        print(f"Fill {i+1} size: {fill} | Client Order executed! Filled: {fill - position} units")
        break

```
