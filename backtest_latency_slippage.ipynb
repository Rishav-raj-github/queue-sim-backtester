# Latency & Slippage Profile
Simulates order execution slippage outcomes under varying round-trip network delays (1ms to 50ms).


```python
import numpy as np
import matplotlib.pyplot as plt

latencies_ms = np.linspace(1, 50, 50)
# Price slippage in Bps scales with round trip time
slippage_bps = 0.15 * latencies_ms + np.random.normal(0, 0.2, 50)

plt.figure(figsize=(10, 4))
plt.scatter(latencies_ms, slippage_bps, color="rose", alpha=0.8)
plt.title("Slippage scale relative to RTT delays")
...

```
